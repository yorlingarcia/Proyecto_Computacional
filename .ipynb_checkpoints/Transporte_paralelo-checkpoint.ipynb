$\textbf{Transporte paralelo}$

$$\frac{D}{d  \lambda}T_{\nu_1, \nu_2, ..., \nu_l}^{~~~~~~~~~~~~~~~\mu_1, \mu_2, ..., \mu_k} = 0$$
$$\frac{D}{d  \lambda} = \frac{dx^{\mu}}{d \lambda} \nabla_{\mu}$$
referencia ------ https://www.preposterousuniverse.com/grnotes/
Definiendo un tensor $V^{\mu}$, su transporte paralelo esta dado por

$$\frac{D}{d  \lambda}V^{\mu} = \frac{dx^{\sigma}}{d \lambda} \nabla_{\sigma}V^{\mu} = \frac{dx^{\sigma}}{d \lambda} \partial_{\sigma}V^{\mu} + \Gamma_{\sigma\rho}^{\mu} \frac{dx^{\sigma}}{d \lambda}V^{\rho}= 0$$

$$ \frac{d}{d \lambda}V^{\mu} + \Gamma_{\sigma\rho}^{\mu} \frac{dx^{\sigma}}{d \lambda}V^{\rho}= 0$$

Suponiendo una conexión levi-civita, la conexión toma la forma

$$ \Gamma_{\mu \nu}^{\rho} = \frac{1}{2}g^{\rho \sigma}\left(\partial_{\mu}g_{\nu\sigma} + \partial_{\nu}g_{\mu\sigma} - \partial_{\sigma}g_{\mu\nu}\right)$$

si $V = 3e_z$, ademas tomando la metrica euclidiana (tomando no evolucion temporal) $\eta_{\mu \nu}$, los valores correspondientes a la conexión es

```python
###########################################
###########################################
import numpy as np
import sympy as sp
from math import sin
from IPython.display import Latex, display
###########################################
###########################################

print('--------metrica euclidiana----------')
D = 2
# r, theta variables simbolicas
metrica = sp.Matrix([['1' , '0' , '0'],['0', '1', '0'],['0', '0', '1']])
print(metrica)
print('-------------------------')

metricainv = metrica.inv()
gamma = np.zeros((D+1,D+1,D+1),list) # simbolos de christoffel
x = ['t','r','theta','phi']

sum = 0
for i in range(D+1): #lambda
    for j in range(D+1): # mu
        for k in range(D+1): # nu
            for l in range(D+1): # sigma
                sum += 0.5*metricainv[i,l]*(sp.diff(metrica[k,l],x[j]) 
                                            + sp.diff(metrica[j,l],x[k]) - sp.diff(metrica[j,k],x[l]))
            gamma[i][j][k] = sum
            sum = 0
aux = 0 # variable suma
for i in range(D+1):
    for j in range(D+1):
        for k in range(D+1): 
            if gamma[i,j,k] != 0: 
                print("gamma(",x[i], x[j], x[k],") = ",gamma[i,j,k])
            else:
                aux += 1
if aux == (D+1)**3:
    print(" los valres de los coeficientes son cero")
```
